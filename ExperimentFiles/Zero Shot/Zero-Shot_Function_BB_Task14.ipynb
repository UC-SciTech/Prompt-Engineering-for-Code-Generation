### BIG-bench\\ c14
Infer and compute a function over lists of natural numbers for a given input given a predetermined set of examples The target function is remove all but elements 3 through 7.

```python
# Task ID: BIG-bench/c14
# Function Generation - ZS
# Model 1

import timeit
def remove_except_3_to_7_model1(input_list):
    # Initialize an empty list to store the filtered elements
    filtered_list = []
    
    # Iterate through the input list
    for num in input_list:
        # Check if the number is within the range 3 to 7 (inclusive)
        if 3 <= num <= 7:
            # If the number satisfies the condition, append it to the filtered list
            filtered_list.append(num)
    
    # Return the filtered list
    return filtered_list

import pandas as pd
mod1_input1 = [2, 5, 1, 8]
mod1_input2 = [4, 1]
mod1_input3 = [1, 3, 9]

mod1_output1 = remove_except_3_to_7_model1(mod1_input1)
mod1_output2 = remove_except_3_to_7_model1(mod1_input2)
mod1_output3 = remove_except_3_to_7_model1(mod1_input3)

# Execution Time
execution_time_mod1_output1 = timeit.timeit(lambda: remove_except_3_to_7_model1(mod1_input1), number=10000)
print("Execution time Model 1 Output 1:", format(execution_time_mod1_output1, '.5f'))

execution_time_mod1_output2 = timeit.timeit(lambda: remove_except_3_to_7_model1(mod1_input2), number=10000)
print("Execution time Model 1 Output 2:", format(execution_time_mod1_output2, '.5f'))
execution_time_mod1_output3 = timeit.timeit(lambda: remove_except_3_to_7_model1(mod1_input3), number=10000)
print("Execution time Model 1 Output 3:", format(execution_time_mod1_output3, '.5f'))

average_time_mod1output = (execution_time_mod1_output1 + execution_time_mod1_output2 + execution_time_mod1_output3) / 3
print("Average Execution Time for Model 1 Output:", format(average_time_mod1output, '.5f'))

def evaluate_list_metrics(expected_output, actual_output):
    true_positives = sum(1 for num in actual_output if num in expected_output)
    false_positives = sum(1 for num in actual_output if num not in expected_output)
    false_negatives = sum(1 for num in expected_output if num not in actual_output)

    precision = true_positives / (true_positives + false_positives) if true_positives + false_positives > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if true_positives + false_negatives > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0

    accuracy = sum(1 for num in actual_output if num in expected_output) / max(len(expected_output), len(actual_output))

    exact_match_accuracy = 1 if expected_output == actual_output else 0

    return {
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1_score,
        "Accuracy": accuracy,
        "Exact Match": exact_match_accuracy
    }


expected_outputs_c14 = [[1, 8], [], [9]]
actual_outputs_mod1c14 = [mod1_output1, mod1_output2, mod1_output3]
print("Expected Output: ", expected_outputs_c14)
print("Actual Output: ", actual_outputs_mod1c14)

metrics1 = [evaluate_list_metrics(expected, actual) for expected, actual in zip(expected_outputs_c14, actual_outputs_mod1c14)]

# Create DataFrame directly from the list of dictionaries
df_metrics1_new = pd.DataFrame(metrics1)

# Calculate the average for precision, recall, F1 score, accuracy, and exact match
avg_metrics1 = df_metrics1_new.mean(axis=0)
avg_metrics1.name = 'Model 1 Average'


# Print the new DataFrame
print(df_metrics1_new)











```
