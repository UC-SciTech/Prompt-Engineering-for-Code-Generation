# Task ID 107
Prompt: Write a Python function that counts even and odd palindromes within a range (1, n), inclusive. It returns a tuple containing the count of even and odd palindromes.

### Revised Prompt for Chain Of Thought
Prompt: We want to find even and odd palindromes within a given range. Let's break this down step-by-step:

1. Identify Palindromes: How can we check if a number is a palindrome? We can convert the number to a string and compare it to its reversed version. If they are equal, it's a palindrome.

2. Separate Even/Odd: Once we identify a palindrome, we need to determine if it's even or odd. We can use the modulo operator (%) to check if the number is divisible by 2 (even) or not (odd).

3. Iterate and Count: We need to iterate through the range from 1 to n (inclusive). For each number, check if it's a palindrome using the method from step 1. Then, based on step 2, increment a counter for even or odd palindromes.

4. Return Results: After iterating through all numbers, we should have two counters: one for even and one for odd palindromes. We can return these counters as a tuple.

```python
# Task ID: HumanEval/107
# Function - CoT
# Model 1

import timeit
import pandas as pd

def is_palindrome(n):
    return str(n) == str(n)[::-1]

def is_even(n):
    return n % 2 == 0

def is_odd(n):
    return not is_even(n)

def even_odd_palindrome1(n):
    even_palindrome_count = 0
    odd_palindrome_count = 0
    
    for i in range(1, n+1):
        if is_palindrome(i):
            if is_even(i):
                even_palindrome_count += 1
            else:
                odd_palindrome_count += 1
                
    return even_palindrome_count, odd_palindrome_count

# Test cases
mod1_output1 = even_odd_palindrome1(123)  # (8, 13)
mod1_output2 = even_odd_palindrome1(12)  # (4, 6)
mod1_output3 = even_odd_palindrome1(3)  # (1, 2)


# Execution Time
execution_time_mod1_output1 = timeit.timeit(lambda: even_odd_palindrome1(123), number=10000)
print("Execution time Model 1 Output 1:", format(execution_time_mod1_output1, '.5f'))

execution_time_mod1_output2 = timeit.timeit(lambda: even_odd_palindrome1(12), number=10000)
print("Execution time Model 1 Output 2:", format(execution_time_mod1_output2, '.5f'))

execution_time_mod1_output3 = timeit.timeit(lambda: even_odd_palindrome1(3), number=10000)
print("Execution time Model 1 Output 3:", format(execution_time_mod1_output3, '.5f'))

average_time_mod1output = (execution_time_mod1_output1 + execution_time_mod1_output2 + execution_time_mod1_output3) / 3
print("Average Execution Time for Model 1 Output:", format(average_time_mod1output, '.5f'))

def evaluate_list_metrics(expected_output, actual_output):
    if isinstance(expected_output, int) or isinstance(actual_output, int):
        # If either expected_output or actual_output is an integer,
        # convert them to lists for comparison
        expected_output = [expected_output]
        actual_output = [actual_output]

    if len(expected_output) != len(actual_output):
        raise ValueError("Expected and actual output lengths must be the same")

    exact_match_accuracy = 1 if expected_output == actual_output else 0

    true_positives = sum(1 for expected, actual in zip(expected_output, actual_output) if expected == actual)
    false_positives = sum(1 for expected, actual in zip(expected_output, actual_output) if expected == 0 and actual == 1)
    false_negatives = sum(1 for expected, actual in zip(expected_output, actual_output) if expected == 1 and actual == 0)

    precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    accuracy = sum(1 for expected, actual in zip(expected_output, actual_output) if expected == actual) / len(expected_output)

    return {
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1_score,
        "Accuracy": accuracy,
        "Exact Match": exact_match_accuracy
    }

expected_outputs = [(8, 13), (4, 6), (1, 2)]
actual_outputs_mod1 = [mod1_output1, mod1_output2, mod1_output3]
print("Expected Output: ", expected_outputs)
print("Actual Output: ", actual_outputs_mod1)

metrics1 = [evaluate_list_metrics(expected, actual) for expected, actual in zip(expected_outputs, actual_outputs_mod1)]

# Create DataFrame directly from the list of dictionaries
df_metrics1_new = pd.DataFrame(metrics1)

# Calculate the average for precision, recall, F1 score, accuracy, and exact match
avg_metrics1 = df_metrics1_new.mean(axis=0)
avg_metrics1.name = 'Model 1 Average'

# Concatenate the average row to the DataFrame
df_metrics1_new = pd.concat([df_metrics1_new, avg_metrics1.to_frame().transpose()])

# Print the new DataFrame
print(df_metrics1_new)

```

```python
# Task ID: HumanEval/107
# Function - CoT
# Model 3

def generate_palindromes(n):
    palindromes = []
    for num in range(1, n + 1):
        num_str = str(num)
        palindrome = int(num_str + num_str[::-1])  # Create a palindrome by appending the reverse of the number
        palindromes.append(palindrome)
    return palindromes

def separate_even_odd(palindromes):
    even_palindromes = [num for num in palindromes if num % 2 == 0]
    odd_palindromes = [num for num in palindromes if num % 2 != 0]
    return even_palindromes, odd_palindromes


def even_odd_palindrome3(n):
    palindromes = generate_palindromes(n)
    even_palindromes, odd_palindromes = separate_even_odd(palindromes)
    return len(even_palindromes), len(odd_palindromes)


# Test cases
mod3_output1 = even_odd_palindrome3(123)  # (8, 13)
mod3_output2 = even_odd_palindrome3(12)  # (4, 6)
mod3_output3 = even_odd_palindrome3(3)  # (1, 2)

# Execution Time
execution_time_mod3_output1 = timeit.timeit(lambda: even_odd_palindrome3(123), number=10000)
print("Execution time Model 3 Output 1:", format(execution_time_mod3_output1, '.5f'))

execution_time_mod3_output2 = timeit.timeit(lambda: even_odd_palindrome3(12), number=10000)
print("Execution time Model 3 Output 2:", format(execution_time_mod3_output2, '.5f'))

execution_time_mod3_output3 = timeit.timeit(lambda: even_odd_palindrome3(3), number=10000)
print("Execution time Model 3 Output 3:", format(execution_time_mod3_output3, '.5f'))

average_time_mod3output = (execution_time_mod3_output1 + execution_time_mod3_output2 + execution_time_mod3_output3) / 3
print("Average Execution Time for Model 3 Output:", format(average_time_mod3output, '.5f'))

def evaluate_list_metrics(expected_output, actual_output):
    exact_match_accuracy = 1 if expected_output == actual_output else 0
    
    # Calculate true positives
    true_positives = sum(1 for i in range(min(len(expected_output), len(actual_output))) if expected_output[i] == actual_output[i])
    
    # Calculate false positives
    false_positives = sum(1 for i in range(len(actual_output)) if actual_output[i] not in expected_output)
    
    # Calculate false negatives
    false_negatives = sum(1 for char in expected_output if char not in actual_output)
    
    # Calculate precision, recall, and F1 score
    precision = true_positives / (true_positives + false_positives) if true_positives + false_positives > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if true_positives + false_negatives > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0

    # Calculate accuracy
    accuracy = sum(1 for i in range(len(expected_output)) if expected_output[i] == actual_output[i]) / max(len(expected_output), len(actual_output))

    return {
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1_score,
        "Accuracy": accuracy,
        "Exact Match": exact_match_accuracy
    }


expected_outputs = [(8, 13), (4, 6), (1, 2)]
actual_outputs_mod3 = [mod3_output1, mod3_output2, mod3_output3]
print("Expected Output: ", expected_outputs)
print("Actual Output: ", actual_outputs_mod3)

metrics3 = [evaluate_list_metrics(expected, actual) for expected, actual in zip(expected_outputs, actual_outputs_mod3)]

# Create DataFrame directly from the list of dictionaries
df_metrics3_new = pd.DataFrame(metrics3)

# Calculate the average for precision, recall, F1 score, accuracy, and exact match
avg_metrics3 = df_metrics3_new.mean(axis=0)
avg_metrics3.name = 'Model 3 Average'

# Concatenate the average row to the DataFrame
df_metrics3_new = pd.concat([df_metrics3_new, avg_metrics3.to_frame().transpose()])

# Print the new DataFrame
print(df_metrics3_new)
```
