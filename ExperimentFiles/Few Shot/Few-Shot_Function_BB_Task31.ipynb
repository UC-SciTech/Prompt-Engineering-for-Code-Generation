### Revised Prompt for FS

### BIG-bench\\ c31

We want to create a function that takes a list of natural numbers and removes elements whichever are equal: the first two elements or the last two.

Example:
""input"": ""[8, 6, 4, 4, 1, 7, 3, 5, 5]"",
""target"": ""[8, 6, 4, 4, 1, 7, 3]""

""input"": ""[7, 7, 5, 1, 9, 4, 3, 2]"",
""target"": ""[5, 1, 9, 4, 3, 2]"""

```python
# Task ID: BIG-bench/c31
# Function Generation - FS
# Model 1

import timeit
def remove_equal_pairs_model1(nums):
    if len(nums) < 4:
        return nums
    
    # Check if the first two elements are equal
    if nums[0] == nums[1]:
        return nums[2:]
    
    # Check if the last two elements are equal
    if nums[-1] == nums[-2]:
        return nums[:-2]
    
    return nums

mod1_input1 = [9, 7, 9, 9] #  [9, 7]]
mod1_input2 = [6, 6, 7, 5, 0, 2, 9] # [7, 5, 0, 2, 9]
mod1_input3 = [8, 4, 4, 4] # [8, 4]

mod1_output1 = remove_equal_pairs_model1(mod1_input1)
mod1_output2 = remove_equal_pairs_model1(mod1_input2)
mod1_output3 = remove_equal_pairs_model1(mod1_input3)

# Execution Time
execution_time_mod1_output1 = timeit.timeit(lambda: remove_equal_pairs_model1(mod1_input1), number=10000)
print("Execution time Model 1 Output 1:", format(execution_time_mod1_output1, '.5f'))

execution_time_mod1_output2 = timeit.timeit(lambda: remove_equal_pairs_model1(mod1_input2), number=10000)
print("Execution time Model 1 Output 2:", format(execution_time_mod1_output2, '.5f'))
execution_time_mod1_output3 = timeit.timeit(lambda: remove_equal_pairs_model1(mod1_input3), number=10000)
print("Execution time Model 1 Output 3:", format(execution_time_mod1_output3, '.5f'))

average_time_mod1output = (execution_time_mod1_output1 + execution_time_mod1_output2 + execution_time_mod1_output3) / 3
print("Average Execution Time for Model 1 Output:", format(average_time_mod1output, '.5f'))

import pandas as pd

def evaluate_list_metrics(expected_output, actual_output):
    true_positives = sum(1 for num in actual_output if num in expected_output)
    false_positives = sum(1 for num in actual_output if num not in expected_output)
    false_negatives = sum(1 for num in expected_output if num not in actual_output)

    precision = true_positives / (true_positives + false_positives) if true_positives + false_positives > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if true_positives + false_negatives > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0

    accuracy = sum(1 for num in actual_output if num in expected_output) / max(len(expected_output), len(actual_output))

    exact_match_accuracy = 1 if expected_output == actual_output else 0

    return {
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1_score,
        "Accuracy": accuracy,
        "Exact Match": exact_match_accuracy
    }
expected_outputs_c031 = [[9, 7], [7, 5, 0, 2, 9], [8, 4]]
actual_outputs_mod1c031 = [mod1_output1, mod1_output2, mod1_output3]
print("Expected Output: ", expected_outputs_c031)
print("Actual Output: ", actual_outputs_mod1c031)

metrics1 = [evaluate_list_metrics(expected, actual) for expected, actual in zip(expected_outputs_c031, actual_outputs_mod1c031)]

# Create DataFrame directly from the list of dictionaries
df_metrics1_new = pd.DataFrame(metrics1)

# Calculate the average for precision, recall, F1 score, accuracy, and exact match
avg_metrics1 = df_metrics1_new.mean(axis=0)
avg_metrics1.name = 'Model 1 Average'


# Print the new DataFrame
print(df_metrics1_new)
```

```python
# Task ID: BIG-bench/c31
# Function Generation - FS
# Model 2

import timeit

def remove_equal_pairs_model2(nums):
    while len(nums) >= 2:
        if nums[:2] == nums[-2:]:
            nums = nums[2:]
        elif nums[0] == nums[1]:
            nums = nums[2:]
        elif nums[-1] == nums[-2]:
            nums = nums[:-2]
        else:
            break
    return nums


mod2_input1 = [9, 7, 9, 9] #  [9, 7]]
mod2_input2 = [6, 6, 7, 5, 0, 2, 9] # [7, 5, 0, 2, 9]
mod2_input3 = [8, 4, 4, 4] # [8, 4]

mod2_output1 = remove_equal_pairs_model2(mod2_input1)
mod2_output2 = remove_equal_pairs_model2(mod2_input2)
mod2_output3 = remove_equal_pairs_model2(mod2_input3)

# Execution Time
execution_time_mod2_output1 = timeit.timeit(lambda: remove_equal_pairs_model2(mod2_input1), number=10000)
print("Execution time Model 2 Output 1:", format(execution_time_mod2_output1, '.5f'))

execution_time_mod2_output2 = timeit.timeit(lambda: remove_equal_pairs_model2(mod2_input2), number=10000)
print("Execution time Model 2 Output 2:", format(execution_time_mod2_output2, '.5f'))

execution_time_mod2_output3 = timeit.timeit(lambda: remove_equal_pairs_model2(mod2_input3), number=10000)
print("Execution time Model 2 Output 3:", format(execution_time_mod2_output3, '.5f'))

average_time_mod2output = (execution_time_mod2_output1 + execution_time_mod2_output2 + execution_time_mod2_output3) / 3
print("Average Execution Time for Model 2 Output:", format(average_time_mod2output, '.5f'))

import pandas as pd
def evaluate_list_metrics(expected_output, actual_output):
    exact_match_accuracy = 1 if expected_output == actual_output else 0
    
    true_positives = sum(1 for i in range(min(len(expected_output), len(actual_output))) if expected_output[i] == actual_output[i])
    false_positives = sum(1 for i in range(min(len(expected_output), len(actual_output))) if expected_output[i] == 1 and actual_output[i] == 0)
    false_negatives = sum(1 for i in range(min(len(expected_output), len(actual_output))) if expected_output[i] == 0 and actual_output[i] == 1)

    precision = true_positives / (true_positives + false_positives) if true_positives + false_positives > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if true_positives + false_negatives > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0

    accuracy = sum(1 for i in range(min(len(expected_output), len(actual_output))) if expected_output[i] == actual_output[i]) / max(len(expected_output), len(actual_output))

    return {
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1_score,
        "Accuracy": accuracy,
        "Exact Match": exact_match_accuracy
    }
expected_outputs_c031 = [[9, 7], [7, 5, 0, 2, 9], [8, 4]]
actual_outputs_mod2c031 = [mod2_output1, mod2_output2, mod2_output3]
print("Expected Output: ", expected_outputs_c031)
print("Actual Output: ", actual_outputs_mod2c031)

metrics2 = [evaluate_list_metrics(expected, actual) for expected, actual in zip(expected_outputs_c031, actual_outputs_mod2c031)]

# Create DataFrame directly from the list of dictionaries
df_metrics2_new = pd.DataFrame(metrics2)

# Calculate the average for precision, recall, F1 score, accuracy, and exact match
avg_metrics2 = df_metrics2_new.mean(axis=0)
avg_metrics2.name = 'Model 2 Average'


# Print the new DataFrame
print(df_metrics2_new)
```
